$b_d=0.7,k_d = 2,m_c=1,b_s=0.8$

它这把滑动摩擦当阻尼器处理感觉怎么恰当，可能它指的是和空气的滑动摩擦。与地面那种滑动摩擦本身只和系数与压力有关，与速度无关。

对$m_c$的受力情况设方程，它一共可分为四个力，阻尼器的力，弹簧的力（加速时它们在推箱子，因为$\dot{x_p} - \dot{x}_{in} \le 0$与
$x_p - x_{in} \le 0$），箱子的动力和不那么正常的摩擦力（加速时前两个力作为推力，这两个则把推力消耗掉达到平衡。）

$$
m_c \ddot{x}_p + b_d (\dot{x_p} - \dot{x}_{in}) + k_d (x_p - x_{in}) + b_s \dot{x}_p = 0 
$$

求拉普拉斯变换后代入数值可得到传递函数：

$$
\frac{X_p(s)}{X_{in}(s)} = \frac{0.7s+2}{s^2+2.8s+2}
$$

```matlab
sys = tf([0.7 2],[1 2.8 2])
```

```matlab
impulse(sys)
```

```matlab
step(sys)
```

```matlab
impulse(sys * tf([1],[1 0]))
```

```matlab
impulse(sys * tf([1],[1 0 0]))
```

```matlab
impulse(sys * tf([1],[1 0 0]), linspace(0,7,100))
```
